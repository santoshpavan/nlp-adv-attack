```python
for i in range(0,tlist.size):
    a = tlist.loc[i]
    d1 = ast.literal_eval(a)
    s = rev.loc[i]
    original_s = s
    target_list = list(d1.keys())
    
    if(len(target_list)==2):
        adj_syn = d1.get(target_list[0])
        verb_syn = d1.get(target_list[1])
        #df.loc[i,'count'] = len(adj_syn)*len(verb_syn)
        for adj in adj_syn:
            for verb in verb_syn:
                s = original_s
                s = s.replace(target_list[0],adj)
                s = s.replace(target_list[1],verb)
                #Write code to check sentiment and store score
                print(s)
                df.loc[i,'mod_review']=s
                
    elif(len(target_list)==1):
        syn = d1.get(target_list[0])
        for word in syn:
            s = original_s
            s = s.replace(target_list[0],word)
            #Write code to check sentiment and store score
            print(s)
            df.loc[i,'mod_review']=s
        
        
    
    
```
